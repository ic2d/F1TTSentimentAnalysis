```python
soup = BeautifulSoup(html_content, 'html.parser')

# Find the table
table = soup.find('table', {'class': 'resultsarchive-table'})

# Extract table header
headers = []
for th in table.find('thead').find_all('th'):
    headers.append(th.text.strip())

# Extract table rows
rows = []
for tr in table.find('tbody').find_all('tr'):
    cells = tr.find_all('td')
    row = []
    for cell in cells:
        # Extract text from cell
        text = cell.get_text(strip=True)
        row.append(text)
    rows.append(row)

# Create DataFrame
df21 = pd.DataFrame(rows, columns=headers)

df21
```

```python
import matplotlib.patches as mpatches

plt.rcParams.update({'font.size': 16})
for i in range(0, len(df_f1_races)):
    date = df_f1_races['race_date'][i]
    race = df_f1_races['Grand Prix'][i]
    # Filter for the current race
    race_data = merged_df[merged_df['race_date'] == date]
    
    if len(race_data) == 0:
        continue
    
    print(f"{race}")

    # Group by 10-minute intervals and count LSTM_pred values
    grouped_counts = race_data[['date', 'LSTM_pred']].groupby(pd.Grouper(key='date', freq='15T'))['LSTM_pred'].value_counts().unstack(fill_value=0)
    
    print(sum(grouped_counts[0]))
    print(sum(grouped_counts[1]))
    
    # Make counts of class 0 negative
    grouped_counts[0] = -grouped_counts[0]

    # Plotting
    ax = grouped_counts.plot(kind='bar', stacked=True, color={0: 'red', 1: 'blue'}, figsize=(12,4))
    hour_labels = grouped_counts.index.strftime('%H:%M')
    # Select labels for each hour
    hourly_ticks = range(0, len(grouped_counts), 6)
    ax.set_xticks(hourly_ticks)
    ax.set_xticklabels(hour_labels[hourly_ticks], rotation=45, ha='right')
    ax.set_xlabel('Time (UTC)')
    ax.set_ylabel('Tweets')
    ax.set_title(f'{race} Grand Prix - {date}')
    
    black_patch = mpatches.Patch(color='red', label='Negative')
    white_patch = mpatches.Patch(color='blue', label='Positive')
    plt.legend(handles=[black_patch, white_patch], loc='upper left', ncol=2)
    
    plt.show()
```
